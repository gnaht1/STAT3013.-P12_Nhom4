```python
import pandas as pd
```

```python
# Input Dataset
df = pd.read_excel('D:/Thái/Phân tích thống kê/STAT3013.-P12_Nhom4/Dataset/Gold_data_filtered.xlsx')  
```

```python
# Mean of the closing price of Gold
df['close'].mean()
```

```python
# Median of the closing price of Gold
df['close'].median()
```

```python
# Max/Min of the closing price of Gold
print(df['close'].max())
print(df['close'].min())
```

```python
# Var of the closing price of Gold
df['close'].var(ddof=1)
```

```python
# Standard Deviation of the closing price of Gold
df['close'].std(ddof=1)
```

```python
# Q3 
df['close'].quantile(0.75)
```

```python
# Q1 
df['close'].quantile(0.25)
```

```python
# IQR 
IQR = df['close'].quantile(0.75) - df['close'].quantile(0.25)
IQR
```

```python
# Upper Limit
upper_limit = df['close'].quantile(0.75) + 1.5*IQR
upper_limit
```

```python
#3.11 Lower Limit
lower_limit  = df['close'].quantile(0.25) - 1.5*IQR
lower_limit 
```

```python
# Skewness
gdp_skewness = df['close'].skew()
gdp_skewness
```

```python
# Kurtosis
gdp_kurtosis = df['close'].kurt()
gdp_kurtosis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 

# Box Plot
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['close'], color='#c7675d', width=0.5, linewidth=2, showfliers=False)
plt.title('Box Plot of Closing Price of Gold (2019 - 2024)')
plt.xlabel('Close')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['close'],color='#c7675d', bins=20, kde=True)
plt.title('Histogram of Closing Price of Gold (2019 - 2024)')
plt.xlabel('Close')
plt.ylabel('Frequency')
plt.show()
```
